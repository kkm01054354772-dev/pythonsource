````python
# 가변 파라메터 
# 파라메터의 개수가 일정하지 않은 경우
def info(*args):
    print(args)

info("wine")
info("wine","35","45","65")
info("35","45","65")
info()
````

````python
# Ex.1 연산기 생성(덧셈, 곱셈)
def sum_mul(choice, *args):
    if choice == "+":
        result=0
        for i in args:
            result+=i
    elif choice == "*":
        result = 1
        for i in args:
            result *= i
    return result


print(sum_mul("+",1,2,3,4,5))
print(sum_mul("*",1,2,3,4,5))
````

````python
# **kwargs : 딕셔너리 형태로 처리 key=value

def args_func1(**kwargs):
    for k,v in kwargs.items():
        print(f"{k} : {v}")

args_func1(name="Park")
args_func1(name="Park", name2 ="Kim", name3 ="Hong", active=True)
args_func1(name="Park", age=25 , title="Title")
````

````python
def args_func2(arg1, arg2=5, *args, **kwargs):
    print(arg1, arg2, args, kwargs)

args_func2(10,20)
args_func2(10,20,"park","kim")
args_func2(10,20,"park","kim",age=14,name="choi")
````

````python
# Ex.2
# 함수 정의: 사칙연산 (+,-,*,//)
def fourRules(num1,num2,op):
    if op == "+":
        return num1+num2
    elif op == "-":
        return num1-num2
    elif op == "*":
        return num1*num2
    elif op == "//":
        return num1//num2
    

fourRules(4,2,"//")
````

````python
# Ex.3 input으로 입력받기
def fourRules(num1, num2, op):
    if op == "+":
        return num1 + num2
    elif op == "-":
        return num1 - num2
    elif op == "*":
        return num1 * num2
    elif op == "//":
        return num1 // num2
    
num1 = int(input("num1 : "))
num2 = int(input("num2 : "))
op = input("op : ")

print(f"{num1} {op} {num2} = {fourRules(num1,num2,op)}")
````

````python
# Ex.4 로또 번호 추출
# 리스트, set

def get_number():
    return random.randrange(1,46)

# 중복 x => Set
lotto =set()
while True:
    lotto.add(get_number())
    
    if len(lotto) >= 6:
        break

print("lotto",lotto)

# 정렬 원한다면 => List

list1 = list(lotto)
list1.sort()
print("lotto",list1)
````

### 람다 함수
- 단일문으로 표현되는 익명 함수
- 1회성으로 함수를 만들어 사용

````python
# 숫자를 받아서 **2 리턴

def square(num):
    return num **2

print(square(5))
````

````python
# 람다 함수
square = lambda num:num**2
square(6)
````

````python
add= lambda x,y:x+y
add(6,9)
````

````python
contents = ['bob','charles','alexander3','teddy']

# 글자 길이에 따른 sort  
# def strlen(s):
#     return len(s)


contents.sort(key=lambda s:len(s))
print(contents)
````

````python
contents = ["bob", "charles", "alexander3", "teddy"]
contents.sort(key=len)
contents
````

````python
# filter, map, reduce
list1 =[1,2,3,6,8,9]

# 짝수 요소만 추출한 후 새로운 리스트 생성하기
evens = []

for i in list1:
    if i % 2 == 0:
        evens.append(i)

evens
````

````python
evens = [i for i in list1 if i % 2 == 0]
evens
````

````python
evens=[]

def even(list1):
    for i in list1:
        if i % 2 == 0:
            evens.append(i)

even([1,2,3,6,8,9])
print(evens)
even([11, 22, 3, 6, 8, 9 , 12, 14])
print(evens)
````

````python
# filter

list1 = [1,2,3,6,8,9]

list(filter(lambda x:x%2==0, list1))
````

````python
# map
nums = [1,2,3,6,8,10,11,12,13,14,15]

list(map(lambda x:x**2, nums))
````
