```python
# Ex.5
# 사칙연산기 ( 숫자 : 2개 , 연산자 : +, -, *, /, //, %)

num1 = int(input("input num1 "))
op= input("input op ")
num2 = int(input("input num2 "))

result = 0
if op == "+":
    result =num1+num2
elif op == "-":
    result =num1-num2
elif op == "*":
    result =num1*num2
elif op == "/":
    result =num1/num2
elif op == "//":
    result =num1//num2
elif op == "%":
    result =num1%num2

print(f"{num1} {op} {num2} = {result}")
```
