```python
# Ex.8
classA = [70,60,55,75,95,90,80,85,100]

# 점수의 합과 평균 구하기
summary = 0
average = 0

for i in classA:
    summary +=i

average = summary / len(classA)

print(f"점수 합계 : {summary} 점수 평균 : {average}")

# sum 함수를 사용하면 더 편하게 구할 수 있음
print(f"점수 합계 : {sum(classA)} 점수 평균 : {sum(classA)/len(classA)}")
```

```python
# 주차장 프로그램
# 5대만 주차 가능
parking_lot = []
top, car_name = 0, "A"

while True :
    no = int(input("[1] 자동차 넣기 | [2] 자동차 빼기 | [3] 종료 "))
    if no <= 3:
        if no ==1:
            if top >= 5 :
                print("주차장 꽉 찼음")
            else:
                parking_lot.append(car_name)
                print(f"{car_name}자동차 들어감. 주차장 상태 ==> {parking_lot}")
                car_name = chr(ord(car_name)+1)
                top += 1
        elif no ==2:
            if top > 0:
                car_name = chr(ord(car_name)-1)
                parking_lot.pop()
                print(f"{car_name}자동차 빠짐. 주차장 상태 ==> {parking_lot}")
                top -= 1
            else:
                print("빠져나갈 자동차가 없음")
        elif no ==3:
            print("프로그램 종료")
            break
    else :
        print("번호를 확인해주세요")
```
